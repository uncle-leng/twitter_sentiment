```python
import matplotlib.pyplot as plt
import numpy as np

# in the order of Sydney, Melbourne, Perth, Brisbane, Canberra
bachelor_rate = [16.9, 16.6, 14.5, 14.5, 21.3]

sydney_score = [0.54729882546096797, 0.55562225177829605, 0.55434277818281641, 0.54796133487836152, 0.54782372396880707, 0.554845645460042, 0.55970741887157205, 0.56212108536321337]
melbourne_score = [0.53967548968418155, 0.55091522227575951, 0.54828670345342534, 0.5425110457282486, 0.53987159459738088, 0.54896827781653401, 0.55302605901445745, 0.5600266744846828]
perth_score = [0.53931504463328428, 0.54831467424061064, 0.54766086543921766, 0.54330956769063343, 0.54043737831310945, 0.54338523684932716, 0.55161234756409527, 0.55370927743023646]
brisbane_score= [0.54587474889091958, 0.55772550712813906, 0.55741615868896899, 0.55073070481813657, 0.55065922922401134, 0.55746584073224248, 0.56109094172870999, 0.5658422751872052]
canberra_score = [0.57644307388673488, 0.58916784348340734, 0.57173502893115213, 0.56799447871326092, 0.56103888427900983, 0.58121564804864345, 0.59623354012220153, 0.59859101670838355]

```

```python
sydney_score = np.average(sydney_score)
melbourne_score = np.average(melbourne_score)
perth_score = np.average(perth_score)
brisbane_score = np.average(brisbane_score)
canberra_score = np.average(canberra_score)

labels = ['Sydney', 'Melbourne', 'Perth', 'Brisbane', 'Canberra']

data_to_plot = [sydney_score, melbourne_score, perth_score, brisbane_score, canberra_score]

y_pos = np.arange(len(labels))

# # Create a figure instance
# fig = plt.figure(1, figsize=(12, 9))



# # Create an axes instance
# ax = fig.add_subplot(111)
# ax_line = fig.add_subplot

# ax2 = ax.twinx()
# ax2.plot(y_pos, bachelor_rate)
# # ax2.set_ylabel('sin', color='r')
# # ax2.tick_params('y', colors='r')

# plt.bar(y_pos, data_to_plot, align='center', width=0.3, alpha=0.5)
# plt.xticks(y_pos, labels)
# plt.ylim([0.5, 0.6])
# plt.ylabel('Sentiment Score')
# plt.xlabel('Cities')
# plt.plot(bachelor_rate, label='Line 2')
# plt.show()


fig, ax1 = plt.subplots(figsize=(12,9))

ax1.bar(y_pos,data_to_plot, align='center', width=0.3, alpha=0.5)
ax1.set_xlabel('cities')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('sentiment sscore', color='b')
ax1.tick_params('y', colors='b')
ax1.set_ylim(0.53,0.6)
#ax1.set_xticklabels(labels)

ax2 = ax1.twinx()
ax2.plot(y_pos, bachelor_rate, 'r-')
ax2.set_ylabel('bachelor degree rate(%)', color='r')
ax2.tick_params('y', colors='r')
ax2.set_xticklabels(labels)

# fig.tight_layout()
plt.xticks(y_pos, labels)
plt.show()
```
